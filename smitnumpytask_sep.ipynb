```python
print ("****** task 2 ******")

import numpy as np

Student_data = {
    "st-1":{"Maths":85,"Urdu":70,"Sindhi":40,"Chemistry":90,"Physics":60},
    "st-2":{"Maths":70,"Urdu":75,"Sindhi":40,"Chemistry":90,"Physics":60},
    "st-3":{"Maths":80,"Urdu":70,"Sindhi":95,"Chemistry":90,"Physics":60},
    "st-4":{"Maths":80,"Urdu":70,"Sindhi":50,"Chemistry":85,"Physics":60},
    "st-5":{"Maths":80,"Urdu":75,"Sindhi":40,"Chemistry":90,"Physics":60}
}

# 1. Max/Min marks in every subject
subjects = ["Maths", "Urdu", "Sindhi", "Chemistry", "Physics"]

print("--- Subject Analysis ---")
for sub in subjects:
    # Get all marks for this subject across all students
    marks = [data[sub] for st, data in Student_data.items()]
    marks_array = np.array(marks)

    print(f"{sub}: Max is {np.max(marks_array)}, Min is {np.min(marks_array)}")

# 2. Max/Min marks for every student
print("\n--- Student Analysis ---")
student_totals = {} # We will save totals here for Task 3

for st_id, scores in Student_data.items():
    # Convert student's scores to a list
    marks_list = list(scores.values())
    marks_array = np.array(marks_list)

    student_totals[st_id] = np.sum(marks_array) # Save for ranking

    print(f"{st_id}: Max mark is {np.max(marks_array)}, Min mark is {np.min(marks_array)}")

# 3. Rank students on total marks
print("\n--- Ranking ---")
# Sort the student totals from highest to lowest
sorted_ranking = sorted(student_totals.items(), key=lambda x: x[1], reverse=True)

for rank, (st_id, total) in enumerate(sorted_ranking, 1):
    print(f"Rank {rank}: {st_id} (Total: {total})")

print("Assignment performed, thanks for allow submit the assignment as late")
print("Submitted please")
```
